### 1.1. Problem description and background
Planning a trip to a foreign city can be daunting. Often equipped with limited budget, time, and experience in of the foreign city, visitors want to make the most of their stay. But how? This is where I would like to help. Using data science, the Toronto neighbourhoods dataset, and FourSquare, I would like to answer the following questions.

## 3. Methodology
In the methodology section, we discuss and describe the exploratory data analysis done, any inferential statistical testing performed, and machine learning models used and why.

### 3.1. Data preparation
Data preparation involved scraping neighbourhoods with postal codes starting with 'M'. This entailed neighbourhoods in Toronto. Boroughs with values 'Not Assigned' were imputed by aggergating these cases with neighbourhoods, whose values were not missing, using identical postal codes.
The prepared Toronto dataset was combined with geospatial coordinates dataset containing latitudes and longitudes.

### 3.2. Exploratory data analysis
Data was visually explored by creating a map of Toronto and superimposing the neighbourhoods on top of the map.
Data was further explored by slicing the dataset and obtaining only boroughts with the string 'Toronto', leading us to focus on the downtown-ish areas.
Simple clustering was also applied to divide these neighbourhoods into areas.

### 3.3. More data preparation and exploration
The venues information was obtained from Foursquare and combined with the sliced datset of Toronto neighbourhoods. Now we are working with Toronto neighbourhoods in the downtown area with knowledge of the venues in those neighbourhoods.
Initial exploration showed a total of 1729 cases.

### 3.4. Machine learning techniques
Clustering was applied throughout the process. Clustering helps us see similarities of elements within a cluster and dissimilarities of elements across cluster. It is also intuitive when used in conjunction with Folium / mapping.
